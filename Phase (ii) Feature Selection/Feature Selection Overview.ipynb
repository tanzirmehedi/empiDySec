```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.transforms import blended_transform_factory

# -----------------------------
# Data
# -----------------------------
features = ["ANOV", "COR", "FLAML", "PSO", "WOA"]

accuracy_sets = [
    [0.92, 0.93, 0.93, 0.92, 0.91],
    [0.84, 0.72, 0.88, 0.88, 0.87],
    [0.75, 0.75, 0.75, 0.75, 0.75],
    [0.76, 0.83, 0.85, 0.83, 0.85],
    [0.98, 0.98, 0.98, 0.98, 0.98],
    [0.73, 0.88, 0.88, 0.87, 0.88]
]

feature_sets = [
    [5, 7, 9, 7, 6],
    [4, 4, 11, 9, 8],
    [2, 4, 2, 3, 4],
    [3, 2, 5, 5, 5],
    [10, 20, 5, 15, 21],
    [4, 10, 8, 5, 6]
]

titles = [
    "Filetop Traces",
    "Opensnoop Traces",
    "Installation Traces",
    "TCP Traces",
    "SysCall Traces",
    "Pattern Traces"
]

x = np.array([0, 1.3, 3.4, 5.5, 6.8])
width = 0.45

# -----------------------------
# Style
# -----------------------------
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 14
})

# -----------------------------
# Create 6 figures
# -----------------------------
fig, axes = plt.subplots(3, 2, figsize=(10, 11))
axes = axes.flatten()

for i, ax1 in enumerate(axes):
    ax2 = ax1.twinx()

    accuracy = accuracy_sets[i]
    selected = feature_sets[i]

    # Bars
    bars1 = ax1.bar(
        x - width/2, accuracy, width,
        color="#4C72B0", label="Accuracy", zorder=3
    )
    bars2 = ax2.bar(
        x + width/2, selected, width,
        color="#DD8452", label="Selected Features", zorder=3
    )

    # Axis labels
    ax1.set_ylabel("Accuracy", labelpad=5)
    ax2.set_ylabel("Selected Features", labelpad=5)

    ax1.set_ylim(0.60, 1.00)
    ax2.set_ylim(0, 25)

    ax1.set_xticks(x)
    ax1.set_xticklabels(features)

    # Ticks
    ax1.tick_params(axis='both', labelsize=14)
    ax2.tick_params(axis='y', labelsize=14)

    # Grid
    ax1.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.5, zorder=0)
    ax2.grid(axis='y', linestyle=':', linewidth=0.9, alpha=0.35, zorder=0)

    # Remove top borders
    ax1.spines['top'].set_visible(False)
    ax2.spines['top'].set_visible(False)

    # Group separators
    for xpos in [2.3, 4.6]:
        ax1.axvline(x=xpos, color='gray', linestyle='-', linewidth=0.9)

    # Category labels only for last row
    if i >= 4:
        trans = blended_transform_factory(ax1.transData, ax1.transAxes)
    
        ax1.text(0.65, -0.15, "Statistical", transform=trans,
                 ha='center', va='top', fontweight='bold', fontsize=13)
    
        ax1.text(3.40, -0.15, "ML/DL", transform=trans,
                 ha='center', va='top', fontweight='bold', fontsize=13)
    
        ax1.text(6.15, -0.15, "Probabilistic", transform=trans,
                 ha='center', va='top', fontweight='bold', fontsize=13)

    # Value labels for accuracy
    for bar in bars1:
        h = bar.get_height()
        y = min(h + 0.005, 0.995)
        ax1.text(
            bar.get_x() + bar.get_width()/2,
            y,
            f"{h:.2f}",
            ha='center',
            va='bottom',
            fontsize=14
        )

    # Value labels for selected features
    for bar in bars2:
        h = bar.get_height()
        y = min(h + 0.6, 24.2)
        ax2.text(
            bar.get_x() + bar.get_width()/2,
            y,
            f"{h}",
            ha='center',
            va='bottom',
            fontsize=14
        )

    # Subplot title
    ax1.set_title(titles[i], fontsize=14, pad=15, fontweight='bold')

# Layout
plt.subplots_adjust(
    bottom=0.08,
    left=0.08,
    right=0.92,
    top=0.95,
    hspace=0.38,
    wspace=0.40
)

plt.savefig("six_feature_selection.png", dpi=600, bbox_inches="tight")
plt.show()
```
